```python
#Import library
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
from sklearn.preprocessing import StandardScaler

#Import dataset
df = pd.read_csv('./dataset_lab2.csv', sep = ',')
```

```python
#How many rows have nan values?
df.isna().any(axis=1).sum()
```

```python
df.loc[pd.isnull(df['ulnal'])]
```

```python
#How many rows have nan values?
df.isnull().any(axis=1).sum()
```

```python
#Nan values were replaced with the mean 
df['huml'].fillna(df['huml'].mean(), inplace = True)
df['humw'].fillna(df['humw'].mean(), inplace = True)
df['ulnal'].fillna(df['ulnal'].mean(), inplace = True)
df['ulnaw'].fillna(df['ulnaw'].mean(), inplace = True)
df['feml'].fillna(df['feml'].mean(), inplace = True)
df['femw'].fillna(df['femw'].mean(), inplace = True)
df['tibl'].fillna(df['tibl'].mean(), inplace = True)
df['femw'].fillna(df['femw'].mean(), inplace = True)
df['tibl'].fillna(df['tibl'].mean(), inplace = True)
df['tibw'].fillna(df['tibw'].mean(), inplace = True)
df['tarl'].fillna(df['tarl'].mean(), inplace = True)
df['tarw'].fillna(df['tarw'].mean(), inplace = True)
```

```python
df.describe()
```

```python
df.drop('id', inplace = True, axis =1)
```

```python
scaler_df = StandardScaler()
np_scaled = scaler_df.fit_transform(df)
```

```python
new_df = pd.DataFrame(np_scaled, columns = ['huml','humw','ulnal','ulnaw','feml','femw','tibl','tibw','tarl','tarw'])
new_df
```

```python
graph = px.scatter_matrix(new_df, dimensions = ['huml','feml','ulnal','tibl', 'tarl','humw','femw','ulnaw','tibw', 'tarw'], width=1100, height=700)
graph.show()
```

```python
new_df.boxplot(column= ['huml','feml','ulnal','tibl', 'tarl','humw','femw','ulnaw','tibw', 'tarw'])
plt.show()
```

```python
new_df.hist(column=['huml','feml','ulnal','tibl', 'tarl','humw','femw','ulnaw','tibw', 'tarw'], bins = 50)
plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range (1,10):
    kmeans_df = KMeans(n_clusters = i, random_state = 0)
    kmeans_df.fit(df)
    wcss.append(kmeans_df.inertia_)
```

```python
wcss
```

```python
graph = px.line( x = range(1,10), y = wcss)
graph.show()
```
